```python
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession

config = ConfigProto()
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)
```

```python
modelo = keras.Sequential([
  #entrada
  keras.layers.Dense(256, activation=tensorflow.nn.relu,input_shape = (18,)),
  #keras.layers.Dense(128,activation=tensorflow.nn.relu),
  #keras.layers.Dense(64,activation=tensorflow.nn.relu),
  keras.layers.Dropout(0.2), #pra evitar overfitting ou underfitting, essa linha significa que 20% vai ser adormecido
  keras.layers.Dense(182, activation=tensorflow.nn.softmax)#saida
])
```

```python
adam = keras.optimizers.Adam(lr=0.001)

#parando_cedo = [keras.callbacks.EarlyStopping(monitor='val_loss'),keras.callbacks.ModelCheckpoint(filepath='melhor_modelo.hdf5',monitor='val_loss',save_best_only=True)]

modelo.compile(optimizer=adam,
               loss='MeanSquaredError', metrics=['accuracy'])
```

```python
historico = modelo.fit(train_array, label_train, batch_size=480,
                       epochs=100)
```
